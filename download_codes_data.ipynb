# Prepare everything needed for demonstration

```python
import subprocess, os
import matplotlib.pyplot as plt
import xarray as xr
```

# Clone Github repo

```python
_ = subprocess.run('git clone -b develop git@github.com:NCAR/GPEP.git', shell=True)
```

# Download the California case

```python
# we set the outpath as "../../GPEP_test_cases"
# two test cases are available: cali2017 (0) and UpCO_SWE (1). here we download "0" which representing California 2017 precipitation and temperature

if os.path.isdir('./GPEP_demo_cases/cali2017'):
    print('Test cases have been downloaded. Skip downloading')
else:
    _ = subprocess.run('python GPEP/tools/get_testcases.py -o ./GPEP_demo_cases -n 0', shell=True)
```

# Create a low-resolution test case to speed up

```python
infile_grid = './GPEP_demo_cases/cali2017/inputs/CALI.gridinfo.0625_v3.nc'
outfile_grid = './GPEP_demo_cases/cali2017/inputs/CALI.gridinfo.5_v3.nc'
ds = xr.load_dataset(infile_grid)
ds = ds.coarsen(x=8, y=8).mean()
ds.to_netcdf(outfile_grid)
```

# Show the study area

```python
fig = plt.figure(figsize=[10, 3])

fig.add_subplot(1, 3, 1)
ds.elev.plot(vmin=0, vmax=3000)

fig.add_subplot(1, 3, 2)
ds.gradient_n_s.plot()

fig.add_subplot(1, 3, 3)
ds.gradient_w_e.plot()

plt.tight_layout()
plt.show()
```
